```python
def getWebText(url): 
    #Download the page using urllib2
    page = urllib2.urlopen(url).read().decode('utf8','ignore')
    #Instantiate soup object
    soup = BeautifulSoup(page,'lxml')
    #from object find all element having tag article, extract text and join together 
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    #Encoded using right paradigm
    return text.encode('ascii', errors='replace').replace("?", " ")

text = getWebText(articleURL)
print text
```

```python
text = getWebText(articleURL)

#Preprocess Downloaded Text
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation

sents = sent_tokenize(text)
#print individual sentences
#print sents

#Convert text into list of words
word_sent = word_tokenize(text.lower())

#Remove stopwords from above word list
customstopwords=set(stopwords.words('english')+list(punctuation))

#Take words which are not in stopwords list
wordslist=[word for word in word_sent if word not in customstopwords]
wordslist
```

```python
#Extracting the most important sentences
from nltk.probability import FreqDist
#Construct a frequency distribution of words
freq = FreqDist(wordslist)
print(freq.most_common)
```

```python
#find top 10 important words in the article excluding stopwords
from heapq import nlargest

print nlargest(10, freq, key=freq.get)

#Compute the score for each sentence (Keys -> Sentences, Values -> Significance Scores)
from collections import defaultdict
#if a key is not present defaultdict will not throw error, it will add the key in the dict

ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]

print ranking

#this will not store in sorted order
top_sent = nlargest(3, ranking, key=ranking.get)
print top_sent

#sort and print in ascending order
print [sents[j] for j in sorted(top_sent)]
```

```python
import urllib2
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.probability import FreqDist
from heapq import nlargest
from collections import defaultdict

def getWebText(url): 
    #Download the page using urllib2
    page = urllib2.urlopen(url).read().decode('utf8','ignore')
    #Instantiate soup object
    soup = BeautifulSoup(page,'lxml')
    #from object find all element having tag article, extract text and join together 
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    #Encoded using right paradigm
    return text.encode('ascii', errors='replace').replace("?", " ")

def summarize_article(text,top_n):
    sents = sent_tokenize(text)
    word_sent = word_tokenize(text.lower())
    customstopwords=set(stopwords.words('english')+list(punctuation))
    wordslist=[word for word in word_sent if word not in customstopwords]
    freq = FreqDist(wordslist)

    ranking = defaultdict(int)

    for i,sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
    top_sent = nlargest(top_n, ranking, key=ranking.get)
    print [sents[j] for j in sorted(top_sent)]

articleURL = "https://timesofindia.indiatimes.com/india/no-role-in-selecting-indian-partners-for-rafale-france/articleshow/65908068.cms"
text = getWebText(articleURL)
summarize_article(text, 3)
```
